```python
#create prices_perc_all
path_data = '/home/nero/Documents/Estudos/DataCamp/Python/Machine_Learning_for_Time_Series_Data_in_Python/datasets/'
import pandas as pd
import numpy as np
```

```python
all_prices = pd.read_csv(path_data + 'prices.csv').pivot(columns = 'symbol', values = 'close', index = 'date')
```

```python
all_prices.head(3)
```

```python
def percent_change(series):
    # Collect all *but* the last value of this window, then the final value
    previous_values = series[:-1]
    last_value = series[-1]

    # Calculate the % difference between the last value and the mean of earlier values
    percent_change = (last_value - np.mean(previous_values)) / np.mean(previous_values)
    return percent_change
```

```python
prices_perc_all = all_prices.rolling(20).aggregate(percent_change)
prices_perc_all.head(3)
```

```python
prices_perc_all_40 = all_prices.rolling(40).aggregate(percent_change)
prices_perc_all_40.head(3)
```

```python
prices_perc_all_40.to_csv(path_data + 'prices_perc_all_40.csv')
```

```python
prices_perc_all.to_csv(path_data + 'prices_perc_all.csv')
```
