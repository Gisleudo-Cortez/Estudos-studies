```python
# exercise 07

"""
Compute relative difference by industry

Computing the relative difference between a company's ratio and its industry average is an excellent way to see if a company is at par with its peers. In this exercise, you'll practice computing the relative difference of the equity multiplier ratio of various FMCG companies from their industry average.

A pandas DataFrame balance_sheet has been loaded for you. pandas has been loaded for you under the alias pd. You can call balance_sheet.columns in the console to assess which columns you can use for the exercise.
"""

# Instructions

"""

    Subset the entries in balance_sheet to get only fmcg companies in the column "comp_type" of balance_sheet.
    Add a column in fcmg where you compute the equity multiplier ratio.
    Add a column to fmcg where you compute the average yearly equity multiplier ratio.
    Add a column to fmcg where you compute the relative difference of the company's equity multiplier ratio with its industry average.

"""

# solution

# Subset the fmcg companies
fmcg = balance_sheet.loc[balance_sheet["comp_type"]=="fmcg"]

# Compute equity multiplier ratio
fmcg.loc[:,"equity_multiplier_ratio"] = fmcg.loc[:,"Total Assets"] / fmcg.loc[:,"Total Stockholder Equity"]

# Compute average equity multiplier ratio by year
fmcg.loc[:,"average_equity_multiplier_ratio"] = fmcg.groupby("Year")["equity_multiplier_ratio"].transform("mean")

# Compute the relative difference
fmcg.loc[:,"relative_difference"] = (fmcg.loc[:,"equity_multiplier_ratio"] / fmcg.loc[:,"average_equity_multiplier_ratio"]) - 1

print(fmcg[["Year", "company", "relative_difference"]])

#----------------------------------#

# Conclusion

"""
Nice job! This is one of the first steps in comparing a company's performance to its peers operating in the same industry. Some companies are doing better than average, and some are worse. However, this is just one ratio, and comparing companies based on just one ratio is usually not good practice. We need to compare different ratios from different families of ratios to get a better picture of which company is doing better.
"""
```



    The first line of code subsets the balance_sheet DataFrame to only include rows where the comp_type is fmcg. This is done using the .loc accessor with a boolean condition. The result is stored in the fmcg DataFrame.

    The second line of code calculates the equity multiplier ratio for each company in the fmcg DataFrame. The equity multiplier ratio is calculated as the total assets divided by the total stockholder equity. The result is stored in a new column in the fmcg DataFrame called equity_multiplier_ratio.

    The third line of code calculates the average equity multiplier ratio for each year. This is done by grouping the fmcg DataFrame by the Year column and then applying the mean function to the equity_multiplier_ratio column. The result is stored in a new column in the fmcg DataFrame called average_equity_multiplier_ratio.

    The fourth line of code calculates the relative difference of each company's equity multiplier ratio with the average equity multiplier ratio for that year. This is done by dividing the equity_multiplier_ratio by the average_equity_multiplier_ratio and then subtracting 1. The result is stored in a new column in the fmcg DataFrame called relative_difference.

    The final line of code prints the Year, company, and relative_difference columns of the fmcg DataFrame. This gives the relative difference of each company's equity multiplier ratio with the industry average for each year.


```python
# exercise 08

"""
Compute ratios by industry

In this exercise, you will assess how the current ratio and debt-to-equity ratio of Tech and FMCG companies are different on average and over time.

An industry's ratio, on average gives us an overall idea of the industry's financial condition. On the other hand, an industry's ratio over timehelps us understand its trends: is the ratio relatively stable or fluctuating over time? Is it improving or worsening?

balance_sheet has been loaded for you as a pandas DataFrame. The current ratio is provided in a column called curr_ratio, and the debt-to-equity ratio is provided in a column called debt_to_equity. pandas has been loaded with the alias pd.
"""

# Instructions

"""

    Select those companies whose column comp_type is tech or fmcg.
    Use .groupby() to compute the average current ratio and debt-to-equity ratio grouped by company type (denoted by the column "comp_type").
    Use .groupby() to compute the average current ratio and debt-to-equity ratio by both "comp_type" and "Year".

"""

# solution

# Choose the tech and fmcg companies
tech_fmcg = balance_sheet.loc[balance_sheet['comp_type'].isin(['tech','fmcg'])]

# Compute the average debt-to-equity ratio and current ratio by industry
groupby_industry = tech_fmcg.groupby('comp_type')[["current_ratio", "debt_to_equity"]].mean()

# Compute the average debt-to-equity ratio and current ratio by industry and over time
groupby_industry_time = tech_fmcg.groupby(['comp_type','Year'])[["current_ratio", "debt_to_equity"]].mean()

print(groupby_industry)
print(groupby_industry_time)

#----------------------------------#

# Conclusion

"""
Well done! Analyzing an industry's financial ratios over time is a great way of seeing the industry trend. You can see that the current ratio of tech companies is consistently falling. You would have missed this if you had just seen the average current ratio. You can also see that the debt-to-equity ratio of FMCG companies in the dataset is very volatile.
"""
```
